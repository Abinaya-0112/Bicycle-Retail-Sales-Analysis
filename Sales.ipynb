```python
sales_data = pd.merge(orders, order_items, on="order_id")
```

```python
sales_data.head()
```

```python
sales_data.info()
```

```python
sales_data["order_date"] = pd.to_datetime(sales_data["order_date"])
```

```python
snapshot_date = sales_data["order_date"].max() + pd.Timedelta(days=1)
```

```python
rfm = sales_data.groupby("customer_id").agg({
    "order_date": lambda x: (snapshot_date - x.max()).days,
    "order_id": "nunique",
    "Total_price": "sum"
})
```

```python
rfm.columns = ["Recency", "Frequency", "Monetary"]
```

```python
rfm.head()
```

```python
rfm = rfm.reset_index()
```

```python
rfm.to_sql(
    "customer_segments",
    con=engine,
    if_exists="replace",
    index=False
)
```
